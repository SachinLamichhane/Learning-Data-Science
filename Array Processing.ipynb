```python
# numpy where

A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
```

```python
# creating a Boolean Array
condition = np.array([True,True, False, False])
```

```python
# Pick a A value when condition is True otherwise choose B value -- zip function allows you to input multiple arrays --list comprehension
answer =  [(A_val if condition else B_val) for A_val, B_val , condition in zip(A,B,condition)]
answer 
```

```python
# np.where shows the values where the condition is True
answer2 = np.where(condition, A,B) 
```

```python
answer2
```
